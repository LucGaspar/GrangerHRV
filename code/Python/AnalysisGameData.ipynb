```python
import pandas as pd
import statsmodels.tsa.stattools as stats
import numpy as np
import glob
from random import randrange
```

```python
filesPath = glob.glob('../Time Series/GameData Tests/*.csv')
```

```python
def bootstrap(x, y, delta, B):

    T = len(x)   
    count = 0
    
    # Passo 0
    corr = x.corr(y, method = 'spearman')
    
    # Passo 8
    for k in range(B):

        print(f'{k+1}/{B}', end='\r')
        
        x_ = pd.Series()
        y_ = pd.Series()

        # Passo 3
        while len(x_) < T:
            # Passo 1
            i = randrange(T - delta)

            # Passo 2
            bloco = x[i : i + delta]
            x_ = pd.concat([x_, bloco])

        # Passo 6
        while len(y_) < T:
            # Passo 4
            j = randrange(T - delta)

            # Passo 5
            bloco = y[j : j + delta]
            y_ = pd.concat([y_, bloco])

        # Passo 7
        corr_ = x_.corr(y_, method = 'spearman')

        # Passo 9_1
        if (abs(corr_) > abs(corr)):
            count = count + 1
            
    # Passo 9_2
    p_value = count / float(B)
    
    return p_value
```
