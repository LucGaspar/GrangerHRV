```python
import pandas as pd
import statsmodels.tsa.stattools as stats
import numpy as np
import glob
```

```python
filesPath = glob.glob('../Time Series/Testes em 100/*.csv')
```

# A x B

## Parte 1

```python
Granger = pd.DataFrame(columns = ['windowSize (ms)', 'Modo', 'Direction', 'Lag', 'Chi2', 'p'])
timeSeriesM = pd.read_csv(filesPath[0])
timeSeriesS = pd.read_csv(filesPath[1])
timeSeriesM = timeSeriesM[timeSeriesM.index < timeSeriesM.shape[0] / 2]
timeSeriesS = timeSeriesS[timeSeriesS.index < timeSeriesS.shape[0] / 2]
Z1 = timeSeriesM[['Pulse X', 'Pulse Y']].to_numpy(dtype = bool)
Z2 = timeSeriesM[['Pulse Y', 'Pulse X']].to_numpy(dtype = bool)
Z3 = timeSeriesS[['Pulse X', 'Pulse Y']].to_numpy(dtype = bool)
Z4 = timeSeriesS[['Pulse Y', 'Pulse X']].to_numpy(dtype = bool)
```

```python
timeSeriesM['Pulse X'] = timeSeriesM['Pulse X'].map({False : 0, True : 1})
timeSeriesM['Pulse Y'] = timeSeriesM['Pulse Y'].map({False : 0, True : 1})
timeSeriesS['Pulse X'] = timeSeriesS['Pulse X'].map({False : 0, True : 1})
timeSeriesS['Pulse Y'] = timeSeriesS['Pulse Y'].map({False : 0, True : 1})
```
