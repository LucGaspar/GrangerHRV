```python
import pandas as pd
import statsmodels.tsa.stattools as stats
import numpy as np
import glob
from random import randrange
```

# RR

```python
dfM = pd.read_csv('../Data/Processed/1.csv').loc[0:10000, :]
dfS = pd.read_csv('../Data/Processed/2.csv').loc[0:10000, :]

Z1 = dfM[['Pulse A', 'Pulse B']].to_numpy(dtype = int)
Z2 = dfM[['Pulse B', 'Pulse A']].to_numpy(dtype = int)

Z3 = dfS[['Pulse A', 'Pulse B']].to_numpy(dtype = int)
Z4 = dfS[['Pulse B', 'Pulse A']].to_numpy(dtype = int)

Granger = pd.DataFrame(columns = ['Modo', 'Direction', 'Lag', 'Chi2', 'p'])
result1 = stats.grangercausalitytests(Z1, 17, addconst=True, verbose=True)
result2 = stats.grangercausalitytests(Z2, 17, addconst=True, verbose=True)
result3 = stats.grangercausalitytests(Z3, 17, addconst=True, verbose=True)
result4 = stats.grangercausalitytests(Z4, 17, addconst=True, verbose=True)

for i in range(0, 17):
    delaySon = i + 1
    ssr_ftest = result1[delaySon][0]['ssr_ftest']
    ssr_qtest = result1[delaySon][0]['ssr_chi2test']
    p = ssr_qtest[1]
    q = ssr_qtest[0]
    Granger = Granger.append([{'Direction' : 'A->B', 'Modo': 'M', 'Lag' : delaySon, 'Chi2': q, 'p' : p}])
    
    ssr_ftest = result3[delaySon][0]['ssr_ftest']
    ssr_qtest = result3[delaySon][0]['ssr_chi2test']
    p = ssr_qtest[1]
    q = ssr_qtest[0]
    Granger = Granger.append([{'Direction' : 'A->B', 'Modo': 'S', 'Lag' : delaySon, 'Chi2': q, 'p' : p}])
    
    ssr_ftest = result2[delaySon][0]['ssr_ftest']
    ssr_qtest = result2[delaySon][0]['ssr_chi2test']
    p = ssr_qtest[1]
    q = ssr_qtest[0]
    Granger = Granger.append([{'Direction' : 'B->A', 'Modo': 'M', 'Lag' : delaySon, 'Chi2': q, 'p' : p}])
    
    ssr_ftest = result4[delaySon][0]['ssr_ftest']
    ssr_qtest = result4[delaySon][0]['ssr_chi2test']
    p = ssr_qtest[1]
    q = ssr_qtest[0]
    
    Granger = Granger.append([{'Direction' : 'B->A', 'Modo': 'S', 'Lag' : delaySon, 'Chi2': q, 'p' : p}])
    
GrangerAB = Granger.copy()
GrangerAB.round(2).to_csv('../Data/Processed/Results/Pulse/AB.csv', index = None)
```
