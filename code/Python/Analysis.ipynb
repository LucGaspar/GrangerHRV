```python
import pandas as pd
import statsmodels.tsa.stattools as stats
```

```python
import numpy as np
import glob
```

```python
filesPath = glob.glob('../Time Series/Test/*.csv')
Granger = pd.DataFrame(columns = ['windowSize (ms)', 'Delay', 'F', 'Chi2', 'p'])

for path in filesPath:
    timeSeries = pd.read_csv(path)
    groupedBy = int(path.split('_')[1][:-4])
    delayDad   = int(5 * 100 / groupedBy)

    Z = timeSeries[['Pulse X', 'Pulse Y']].to_numpy(dtype = bool)

    result = stats.grangercausalitytests(Z, delayDad, addconst=True, verbose=True)

    for i in range(delayDad):
        delaySon = i + 1
        ssr_ftest = result[delaySon][0]['ssr_ftest']
        ssr_qtest = result[delaySon][0]['ssr_chi2test']
        f = ssr_ftest[0]
        p = ssr_ftest[1]
        q = ssr_qtest[0]
        Granger = Granger.append([{'windowSize (ms)' : groupedBy, 'Delay' : delaySon, 'F' : f, 'Chi2': q, 'p' : p}])
        
Granger0 = Granger.copy()
```
