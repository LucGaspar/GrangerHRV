```python
import pandas as pd
import statsmodels.tsa.stattools as stats
```

```python
import numpy as np
import glob
```

```python
x = pd.read_csv('X.csv').transpose().reset_index()['index']
y = pd.read_csv('Y.csv').transpose().reset_index()['index']
```

```python
x = pd.Series(x)
y = pd.Series(y)
```

```python
timeSeries = pd.DataFrame({'X' : x, 'Y' : y})
```

```python
timeSeries
```

```python
Z = timeSeries[['X', 'Y']].to_numpy(dtype = float)
```

```python
stats.grangercausalitytests(Z, 10, addconst=True, verbose=True)
```
