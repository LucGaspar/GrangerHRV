```python
def funcResult(row, columns):
    if (row[columns[-2]] > row[columns[-3]]):
        return columns[-2]
    elif (row[columns[-2]] == row[columns[-3]]):
        return 'Empate'
    else:
        return columns[-3]
```

```python
for k in range(len(paths)):
    p = pd.read_csv(paths[k][0]).dropna().reset_index(drop = True)
    p['Time'] = paths[k][1]
    p['Time'] = pd.to_datetime(p['Time'], format = "%Y/%m/%d %H:%M:%S.%f")
    p['Time'] = p.apply(lambda row : row.Time + datetime.timedelta(milliseconds = row.TicksTime - paths[k][2]), axis = 1)
    p = p.drop(columns = ['TicksTime'])
    p['Resultado'] = p.apply(lambda row : funcResult(row, p.columns), axis = 1)

    p['Attacking'] = None

    lastX = None
    for i, x in p.BallPosX.iteritems():
        print(f'{i+1}/{p.shape[0]}', end='\r')

        if (lastX == None):
            lastX = x
            continue

        if (x > lastX):
            p['Attacking'][i] = paths[k][5]
        else:
            p['Attacking'][i] = paths[k][6]

        lastX = x

    pos1Text = p.columns[3]
    pos2Text = p.columns[4]

    newColumn = pos1Text + 'UpDown'

    p[newColumn] = None

    lastY = None

    for i, y in p[pos1Text].iteritems():
        print(f'{i+1}/{p.shape[0]}', end='\r')

        if (lastY == None):
            lastY = y
            continue

        if (y == lastY):
            p[newColumn][i] = 'Stop'
        elif (y > lastY):
            p[newColumn][i] = 'Down'
        else:
            p[newColumn][i] = 'Up'

        lastY = y

    newColumn = pos2Text + 'UpDown'

    p[newColumn] = None

    lastY = None

    for i, y in p[pos2Text].iteritems():
        print(f'{i+1}/{p.shape[0]}', end='\r')

        if (lastY == None):
            lastY = y
            continue

        if (y == lastY):
            p[newColumn][i] = 'Stop'
        elif (y > lastY):
            p[newColumn][i] = 'Down'
        else:
            p[newColumn][i] = 'Up'

        lastY = y

    p = p[(p.Time >= paths[k][3]) & (p.Time <= paths[k][4])]
    p = p.reset_index(drop = True)

    p.to_csv(paths[k][0] + '_NewFeatures', index = None)
```
