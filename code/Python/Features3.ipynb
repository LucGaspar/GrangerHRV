```python
import pandas as pd
import glob
```

```python
filesPath = glob.glob('../Time Series/Data/*.csv')
#newFilesPath = glob.glob('../Time Series/Data/100/*.csv')
newFilesPath = glob.glob('../Time Series/Data/15/*.csv')
```

```python
window = 15
```

```python
for i in range(len(filesPath)):
    #print(f'{i+1}/{len(filesPath)}', end='\r')
    
    path      = filesPath[i]
    newPath   = newFilesPath[i]
    timeSerie = pd.read_csv(path, parse_dates = ['Time'])
    
    size   = timeSerie.shape[0]
    
    numericColumns = timeSerie.select_dtypes(float).columns
    objectColumns  = timeSerie.select_dtypes(object).columns
    pulseColumns   = timeSerie.select_dtypes('int64').columns
    timeColumns    = timeSerie.select_dtypes('datetime64[ns]').columns
      
    numericDataFrame = pd.DataFrame(columns = numericColumns)
    objectDataFrame  = pd.DataFrame(columns = objectColumns)
    pulseDataFrame   = pd.DataFrame(columns = pulseColumns)
    timeDataFrame    = pd.DataFrame(columns = timeColumns)
    
    count  = 0
    inf    = 0
    sup    = 0

    while sup < size:
        print(f'{i+1}/{len(filesPath)} | {sup}/{size}', end='\r')

        inf = window * count; sup = inf + window

        q = timeSerie.loc[inf:sup, objectColumns].mode()
        objectDataFrame = objectDataFrame.append(q, ignore_index = True)
        
        q = timeSerie.loc[inf:sup, timeColumns].max()
        timeDataFrame = timeDataFrame.append(q, ignore_index = True)
        
        q = timeSerie.loc[inf:sup, numericColumns].mean()
        numericDataFrame = numericDataFrame.append(q, ignore_index = True)
        numericDataFrame = numericDataFrame.round(0)
        
        q = timeSerie.loc[inf:sup, pulseColumns].max()
        pulseDataFrame = pulseDataFrame.append(q, ignore_index = True)

        count = count + 1
    
    dataset = pd.concat([timeDataFrame, pulseDataFrame, numericDataFrame, objectDataFrame], axis = 1)
    dataset.to_csv(newPath, index = None)
```

```python
newFilesPath = glob.glob('../Time Series/Data/15/*.csv')

for j in range(0, 2):
    df = pd.read_csv(newFilesPath[j])
    df = df.fillna('None')

    for i, x in df.iterrows():
        if (df['PosBUpDown'][i] == 'None'):
            if (df['PosB'][i] > df['PosB'][i - 1]):
                df['PosBUpDown'][i] = 'Down'
            elif (df['PosB'][i] == df['PosB'][i - 1]):
                df['PosBUpDown'][i] = 'Stop'
            else:
                df['PosBUpDown'][i] = 'Up'

        if (df['PosAUpDown'][i] == 'None'):
            if (df['PosA'][i] > df['PosA'][i - 1]):
                df['PosAUpDown'][i] = 'Down'
            elif (df['PosA'][i] == df['PosA'][i - 1]):
                df['PosAUpDown'][i] = 'Stop'
            else:
                df['PosAUpDown'][i] = 'Up'

    df = df.replace('None', None)
    df = df.dropna(subset = ['Time'])
    df.to_csv(newFilesPath[j] + '_')
```
